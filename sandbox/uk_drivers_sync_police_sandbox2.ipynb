### Aggregate employee group columns

```python
# Let's do it the dirty way: first get the relevant columns

def get_driver_group_columns():
    for column in p.raw.columns:        
        is_group = 'employee group' in column.lower()
        is_driver_group = 'driver' in column.lower()
        is_test_driver_group = 'test' in column.lower()
        
        if is_group and is_driver_group and not is_test_driver_group:
            yield column
            
def get_all_group_columns():
    for column in p.raw.columns:        
        if 'Employee group' in column:
            yield column

              
driver_group_columns = list(get_driver_group_columns())
pprint(driver_group_columns)
print('total =', len(driver_group_columns))
print()

all_group_columns = list(get_all_group_columns())
pprint(list(all_group_columns))
print('total =', len(all_group_columns))
```

```python
# try to extract the actual group name from the columns

def get_employee_group_names():
    for column in get_employee_group_columns():
        yield column.split(': ')[1]
       
fleet_names = list(get_employee_group_names())
print('total = ', len(fleet_names))
fleet_names
```

```python
from pandas import Series

group_as_generator = get_driver_group_columns()
groups = list(group_as_generator)

p.groups = p.raw[groups]
print(p.groups.shape)
p.groups[:5]
```

```python
# a bit of introspection to see how pandas handles iteration

for i, row in p.groups.iterrows():
    print('Type of row[{i}] object: {type_}'.format(i=i, type_=type(row)))
    print('Length of row[{i}] object: {len_}'.format(i=i, len_=len(row)))

    print()
    
    for j, cell in row.iteritems():
        print('In row[{i}], cell[{j}] = {value} ({type_})'.format(i=i, j=j, type_=type(cell), value=cell))
        print()
        break

    break
```

```python
# okay, what we need to do is make a new column with the fleet name

employee_groups_as_dict = {}

for i, groups in p.groups.iterrows():
    groups_the_driver_belongs_to = groups.dropna()

    fleets = []
    for index, fleet in groups_the_driver_belongs_to.iteritems():
        fleet_name = index.split(': ')[1]
        fleets.append(fleet_name)
    
    fleets_as_string = ', '.join(fleets)
    employee_groups_as_dict.update({i: fleets_as_string})
    
fleets_column = Series(employee_groups_as_dict, name='planday_fleets')
print(fleets_column.value_counts(dropna=False))
print('SUM =', sum(fleets_column.value_counts(dropna=False)))
print('LENGTH =', len(fleets_column.value_counts(dropna=False)))
print()
fleets_column.replace('', None, inplace=True)
print(fleets_column.value_counts(dropna=False))
print('SUM =', sum(fleets_column.value_counts(dropna=False)))
print('LENGTH =', len(fleets_column.value_counts(dropna=False)))
```
